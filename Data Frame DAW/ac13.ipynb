```python
# 0. Importa la llibreria pandas
import pandas as pd
```

```python
# 1. Crea un dataframe a partir del CSV adjunt, fent servir l'importador adient de pandas.
# Pots fer servir tres instruccions
dataFrameMatriculats = pd.read_csv("recursos/Alumnes_matriculats_per_ensenyament_i_unitats_dels_centres_docents_20240322.csv")
```

<h5>Llista de valors únics de variables categòriques</h5>
<p style="font-size:12px;">Les variables categòriques són aquelles que només poden tenir un conjunt discret de valors. És a dir, una variable categòrica es una variable no contínua. Per exemple: la variable 'alçada' d'una persona és contínua; en canvi, la variable 'edat' és discreta.</p>

```python
# 2. Mostra una llista de tots els cursos pels quals hi ha registres al dataframe.
# Fes servir el mètode unique() sobre la columna "Curs".
dataFrameMatriculats['Curs'].unique()
```

```python
# Mostra una llista de tots els ensenyaments que contenen la paraula "WEB" (filtra per "nom ensenyament").
# Com pots veure, l'instrucció s'executa amb error
# ValueError: Cannot mask with non-boolean array containing NA / NaN values
# dataFrameMatriculatsValida=dataFrameMatriculats.dropna(subset=["Nom ensenyament"])
dataFrameMatriculats[dataFrameMatriculats]["Nom ensenyament"].str.contains("WEB").unique()
```

```python
# 3. Elimina del df les fileres que tenen "Nom ensenyament" nul.
# df = df[df["Nom ensenyament"].notna()]
dataFrameMatriculats=dataFrameMatriculats[dataFrameMatriculats["Nom ensenyament"].notna()]
dataFrameMatriculats.loc[dataFrameMatriculats["Nom ensenyament"].isnull()] # Per mostrar el resultat
```

```python
# 4. Ara sí, mostra una llista de tots els ensenyaments que contenen la paraula "WEB" (filtra per "nom ensenyament").
# Fes servir el mètode unique perquè no es repeteixin valors.
dataFrameMatriculats[dataFrameMatriculats["Nom ensenyament"].str.contains("WEB")]["Nom ensenyament"].unique()
```

<h5>Query bàsic</h5>

<p style="font-size:16px">El mètode .query() és fonamental per usar amb comoditat i eficiència el dataframe</p>

```python
# 5. Filtra el dataframe perquè quedin només els registres del curs 2022/2023.
# Per fer-ho, completa la següent query. Recorda que la query ha d'anar entre cometes.
dataFrameMatriculats.query('Curs == "2022/2023"')
```

```python
# 6. Filtra el dataframe perquè quedin només els registres corresponents a l'ensenyament
# "CFPS - DESENVOLUPAMENT D'APLICACIONS WEB".
# Alerta: a dins del query, els noms de columnes compostes han d'anar entre backticks (``)
# Alerta: a dins del query, els strings han d'anar entre cometes simples o dobles.
dataFrameMatriculats.query(" `Nom ensenyament` == \"CFPS - DESENVOLUPAMENT D'APLICACIONS WEB\"")

```

```python
# 6bis. Filtra el dataframe perquè quedin només els registres que continguin "APLICACIONS WEB" a "Nom ensenyament".
# Hauràs de fer servir l'operador .str.contains() a dins de la query.
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("APLICACIONS WEB")')
```

<h5>Queries encadenats</h5>

```python
# 7. Encadena dos query perquè es mostrin els registres corresponents a DAW pel curs 2022/2023.
dataFrameMatriculats.query('Curs == "2022/2023"').query('`Nom ensenyament`.str.contains("APLICACIONS WEB")')
```

<p style="font-size:14px">Dos queries encadenats són equivalents amb un únic query que té una condició composta.</p>
<p style="font-size:12px"><i>Per exemple:</i><br>df.query("col1 > 10").query("col2 < 5")<br><i>retorna el mateix que<br></i>df.query("col1 > 10 and col2 < 5")</p>

```python
#7bis Reescriu la instrucció anterior amb un únic query de condició composta
dataFrameMatriculats.query('`Curs`=="2022/2023" and `Nom ensenyament`.str.contains("APLICACIONS WEB")')
```

<h5>Estadístiques generals de matriculació</h5>

```python
# 8. Mostra el número total de matriculats a Catalunya en ensenyaments de qualsevol tipus durant el curs 2022/2023.
# Per fer-ho, completa l'instrucció.
dataFrameMatriculats.query('Curs == "2022/2023"')["Matrícules. Total"].sum()
```

```python
# 9. Mostra el número total de matriculats a Catalunya en ensenyaments de FP (formació professional) durant 
# el curs 2022/2023.
dataFrameMatriculats.query('`Curs`=="2022/2023" and `Nom ensenyament`.str.contains("CFPS")')["Matrícules. Total"].sum()
```

```python
# 10. Mostra el número total de matriculats a Catalunya en ensenyaments de FP (formació professional) durant
# el curs 2022/2023, agrupats per temàtica. Per fer-ho completa la següent instrucció.
dataFrameMatriculats.query('`Curs`=="2022/2023"').query('`Nom ensenyament`.str.contains("CFPS")').groupby('Temàtica')["Matrícules. Total"].sum()
```

```python
# 11. Completa la instrucció anterior perquè els resultats apareguin ordenats de major a menor nombre de 
# matriculats durant el curs 2022/2023. Hauràs de fer servir el mètode .sort_values() amb el paràmetre 
# 'ascending' convenientment assignat.
dataFrameMatriculats.query('`Curs`=="2022/2023"').query('`Nom ensenyament`.str.contains("CFPS")').groupby('Temàtica')["Matrícules. Total"].sum().sort_values(ascending=False)
```

```python
# 12. Reescriu l'instrucció anterior en format multi-línia
dataFrameMatriculats.query('`Curs`=="2022/2023"')\
    .query('`Nom ensenyament`.str.contains("CFPS")')\
    .groupby('Temàtica')["Matrícules. Total"]\
    .sum().sort_values(ascending=False)
```

```python
# 13. Completa la instrucció anterior perquè només apareguin les deu temàtiques amb més matriculacions durant 
# el curs 2022/2023.
# Per fer, pots afegir l'slice [:10] al final de la instrucció.
dataFrameMatriculats.query('`Curs`=="2022/2023"')\
    .query('`Nom ensenyament`.str.contains("CFPS")')\
    .groupby('Temàtica')["Matrícules. Total"]\
    .sum().sort_values(ascending=False).iloc[:10]
```

```python
# 14. Encadena a la instrucció anterior el mètode .plot per graficar les dades en un gràfic de tarta ('pie').
dataFrameMatriculats.query('`Curs`=="2022/2023"')\
    .query('`Nom ensenyament`.str.contains("CFPS")')\
    .groupby('Temàtica')["Matrícules. Total"]\
    .sum()\
    .sort_values(ascending=False)\
    .iloc[:10]\
    .plot(kind='pie',figsize=(20,10),title="Matrícules Total per Temàtica")
```

<h5>Estadístiques de matriculació de DAW</h5>

```python
# 15. Mostra el número total de matriculats a Catalunya a DAW durant el curs 2022/2023.
# Per fer-ho, corregeix la sintaxi de la següent columna
dataFrameMatriculats.query('`Curs`=="2022/2023" and `Nom ensenyament`.str.contains("APLICACIONS WEB")')["Matrícules. Total"].sum()
```

```python
# 16. Mostra en un gràfic de línea l'evolució del número de matriculats al DAW entre els cursos 2015/2016 i 2022/2023.
# Per fer-ho completa la isntrucció.
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")').groupby('Curs')["Matrícules. Total"].sum().plot(figsize=(20,10), kind='line', title='Evolució del número de matriculats al DAW entre els cursos 2015/2016 i 2022/2023', grid=True, fontsize=15, color='green', marker='o')
```

```python
# 17. Reescriu la instrucció anterior en format multi-línia
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
.groupby('Curs')["Matrícules. Total"]\
.sum()\
.plot(figsize=(20,10), kind='line'\
    , title='Evolució del número de matriculats al DAW entre els cursos 2015/2016 i 2022/2023'\
    , grid=True, fontsize=15, color='green', marker='o')
```

```python
# 18. Modifica la instrucció anterior perquè es mostrin al gràfic, enlloc de les matrícules totals, 
# les matrícules d'homes i de dones per separat.
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
.groupby('Curs')[["Matrícules. Homes", "Matrícules. Dones"]]\
.sum()\
.plot(figsize=(20,10), kind='line'\
    , title='Evolució del número de matriculats al DAW entre els cursos 2015/2016 i 2022/2023'\
    , grid=True, fontsize=15, marker='o')
```

```python
# 19. Crea un nou dataframe 'dfDaw' que només contingui els registres pel DAW, agrupats per curs,
# únicament amb les columnes de matrícules totals, matrícules d'homes i matrícules de dones
# Completa la següent instrucció
dfDaw = dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")') \
.groupby('Curs') \
[["Matrícules. Total","Matrícules. Homes","Matrícules. Dones"]].sum()

dfDaw # Per mostrar el resultat
```

```python
# 20. Afegeix al dataframe 'dfDaw' una nova columna anomenada 'Percentatge dones' en que aparegui,per cada curs, 
# el percentatge de dones matriculades respecte del total de matriculats. 
dfDaw["Percentatge dones"] = dfDaw["Matrícules. Dones"] / dfDaw["Matrícules. Total"] * 100
```

```python
# 21. Mostra els valors de la columna 'Percentatge dones', arodonits a un decimal i amb unitats de percentatge (%).
# Fer servir els mètodes .round(), .astype(str), i l'operador de concatenació de cadenes '+'.
dfDaw['Percentatge dones'].round(2).astype(str) + '%'
```

```python
# 22. Mostra el total de matriculats durant el curs 2022/2023 agrupats per cada delegació territorial.
dataFrameMatriculats.query('`Curs` == "2022/2023"').groupby("Nom delegació")["Matrícules. Total"].sum()
```

```python
# 23. Mostra el número total de centres educatius que imparteixen DAW a cada municipi, durant el curs 2022/2023.
#  Hi han d'aparèixer només els 20 primers resultats.
dataFrameMatriculats.query('`Curs` == "2022/2023"')\
.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
.groupby("Nom municipi")["Denominació completa"].count().sort_values(ascending=False).head(20)
```

```python
# 24. Mostra, sense repeticions, el nom dels centres educatius que imparteixen DAW al municipi de Barcelona 
# durant el curs 2022/2023.
# Completa la següent instrucció
dataFrameMatriculats.query('`Curs` == "2022/2023"')\
    .query("`Nom municipi`=='Barcelona'")\
    .query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')["Denominació completa"].unique()
```

```python
# 25. Mostra els 10 centres educatius amb més matriculats a DAW  durant el curs 2022/2023.
#  Hi han d'aparèixer al costat de la "Denominació completa" de cada centre, el municipi al qual pertany.
dataFrameMatriculats.query('`Curs` == "2022/2023"')\
    .query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
    .groupby(["Denominació completa","Nom municipi"])["Matrícules. Total"]\
    .sum().sort_values(ascending=False).head(10)
```

```python
# EXTRA1 Mostrar en un mateix gràfic l'evolució de places públiques i privades pel CFGS de DAW durant el
# període 2015/2016 a 2022/2023
# S'ha de resoldre la consulta en una única instrucció
# Es probable que necessitis fer servir el mètode .unstack()
# dataFrameMatriculatsValida=dataFrameMatriculats[["Curs","Nom naturalesa","Matrícules. Total","Nom ensenyament"]]
# # dataFrameMatriculatsValida.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
#     .groupby(["Curs","Nom naturalesa"])["Matrícules. Total"].sum().unstack().plot(kind="bar",figsize=(12,6))
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
     .groupby(["Curs","Nom naturalesa"])["Matrícules. Total"].sum().unstack().plot(kind="bar",figsize=(12,6))
```

```python
# EXTRA2 Mostrar en una gràfic la distribució geogràfica dels centres que imaprteixen DAW
# Fes servir el tipus de gràfic 'scatter'.
dataFrameMatriculats.query('`Nom ensenyament`.str.contains("DESENVOLUPAMENT D\'APLICACIONS WEB")')\
    .plot(kind='scatter', x='Coordenades GEO X', y="Coordenades GEO Y", title='Distribució geogràfica dels centres que imparteixen DAW',fontsize=8,figsize=(10,8),color='g')
```

```python
# EXTRA3: Mostrar EN UNA SOLA INSTRUCCIÓ l'ensenyament de FP amb més matriculats a cada delegació territorial (2022/2023)
# PISTES:
# Pas 1: filtrar per curs
# Pas 2: filtrar per codi d'estudis (FP)
# Pas 3: fer una agrupació per delegació i ensenyament
# Pas 4: sumar la columna de matrícules totals
# Pas 5: desapilar (unstack) l'índex compost
# Pas 6: treure l'índex (=nom de columna) que correspon al valor màxim de cada fila (idmxax)
dataFrameMatriculats.query('`Curs`=="2022/2023"') \
                   .query('`Nom ensenyament`.str.contains("CFPS")') \
                   .groupby(['Nom delegació','Nom ensenyament'])["Matrícules. Total"] \
                   .sum() \
                   .unstack() \
                   .idxmax(axis=1)

```
